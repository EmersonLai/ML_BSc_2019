# 3-Layer Perceptron

A ideia é representar a área de um círculo unitário com a união composta de ANDs e ORs de n perceptrons. A imagem abaixo ilustra a representação com $n = 4$ perceptrons:

![title](circulocomperceptrons.png)

Com esses $n = 4$ perceptrons e o plano em $R^2$, basta fazer uma operação de AND juntando todos os perceptrons, ou seja: 

$$f = sign(sign(w^T_1x)+sign(w^T_2x)+sign(w^T_3x)+sign(w^T_4x)-3.5)$$

já que queremos $f = 1$ se e só se $sign(w^T_1x)+sign(w^T_2x)+...+sign(w^T_nx) = n$.

Assim, a representação em grafo da fórmula explícita de $f$ é:

![title](multilayer.png)

Vale lembrar que para o caso com $n = 4$ poderíamos fazer AND(AND(P1, P2), AND(P3, P4)), mas também podemos juntar todos os ANDs e tirar uma camada, ficando com duas camadas. Preferi fazer assim.

Assim, seja $W$ a matriz que contém os perceptrons e $x\_a = [x_1; x_2]$: 

```julia
function f(W, x_a)
    x = append!([1.0], x_a)
    h1 = transpose(W)*x
    h2 = append!([1], [sign(h1[i]) for i in 1:length(h1)])
    h3 = sign(transpose(append!([0.5-size(W,2)], [1 for i in 1:(length(h2)-1)]))*h2)
    return h3
end
```

## Exemplo

```julia
using Plots
import Distributions: Uniform
```

Sejam os perceptrons, sendo que em verde é a parte positiva e em vermelho a negativa:

```julia
plot([-2; 2], [1; 1])
plot!(Shape(-2 .+ [0,4,4,0], 1 .+ [0,0,2,2]), opacity=.5)
plot!(Shape(-2 .+ [0,4,4,0], -1 .+ [0,0,2,2]), opacity=.5)
```

```julia
plot([-1; -1], [2; -2])
plot!(Shape(-1 .+ [0,1,1,0], -2 .+ [0,0,4,4]), opacity=.5)
plot!(Shape(-2 .+ [0,1,1,0], -2 .+ [0,0,4,4]), opacity=.5)
```

```julia
plot([-2; 2], [-1; -1])
plot!(Shape(-2 .+ [0,4,4,0], -1 .+ [0,0,1,1]), opacity=.5)
plot!(Shape(-2 .+ [0,4,4,0], -2 .+ [0,0,1,1]), opacity=.5)
```

```julia
plot([1; 1], [-2; 2])
plot!(Shape(1 .+ [0,2,2,0], -2 .+ [0,0,4,4]), opacity=.5)
plot!(Shape(-1 .+ [0,2,2,0], -2 .+ [0,0,4,4]), opacity=.5)
```

Queremos então a região:

```julia
plot([-2; 2], [1; 1])
plot!(Shape(-2 .+ [0,4,4,0], -2 .+ [0,0,4,4]), opacity=.5)
plot!(Shape(-1 .+ [0,2,2,0], -1 .+ [0,0,2,2]), opacity=.6)
plot!([-1; -1], [2; -2])
plot!([-2; 2], [-1; -1])
plot!([1; 1], [-2; 2])
```

W é a matriz com os perceptrons:

```julia
w = [[1 1 1 1];[-1 0 1 0];[0 -1 0 1]]
```

Definimos $F$ daddo w:

```julia
F(x) = f(w, x)
```

Então colocamos pontos aletórios dado o valor de $F$:

```julia
n = 1000
X = rand(Uniform(-2,2), n)
Y = rand(Uniform(-2,2), n)
mais_x = []
mais_y = []
menos_x = []
menos_y = []
for i in 1:n
    if F([X[i], Y[i]]) == 1
        append!(mais_x, X[i])
        append!(mais_y, Y[i])
    else
        append!(menos_x, X[i])
        append!(menos_y, Y[i])
    end
end
plot([-2; 2], [1; 1])
scatter!(menos_x, menos_y)
scatter!(mais_x, mais_y)
plot!([-1; -1], [2; -2])
plot!([-2; 2], [-1; -1])
plot!([1; 1], [-2; 2])
```
