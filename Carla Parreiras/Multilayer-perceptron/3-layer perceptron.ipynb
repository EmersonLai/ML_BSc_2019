 Implementation of 3-layer and n perceptrons (with n=4)

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def Perceptron_3layer_n4(v):
    x=np.array([1,v[0],v[1]])
    n_1=np.array([1,0,-1])
    n_2=np.array([1,0,1])
    n_3=np.array([1,-1,0])
    n_4=np.array([1,1,0])
    
#Classification:
    s_1=np.sign(np.matmul(n_1.T,x))
    s_2=np.sign(np.matmul(n_2.T,x))
    s_3=np.sign(np.matmul(n_3.T,x))
    s_4=np.sign(np.matmul(n_4.T,x))

    f=np.sign(-3.5+s_1+s_2+s_3+s_4)
    return f
```

```python
x = np.random.uniform(-2,2,200)
y = np.random.uniform(-2,2,200)

result=[Perceptron_3layer_n4(p) for p in list(zip(x,y))]
plt.scatter(x,y,c=result)
```
